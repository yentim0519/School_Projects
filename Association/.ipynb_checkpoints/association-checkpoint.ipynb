## 跑apriori演算法

```python
#依照編號（依照日期、客戶分的編號）分類 -> 貨品名稱
pro_list = []
group1 = store_data.groupby(['客戶簡稱', '單據日期'])
```

```python
# 為了github乾淨，沒按開來
group1.groups
```

```python
# 將每位客人的每次消費轉換成list
for i in group1: 
    pro_list.append(i[1]['貨品名稱'].tolist())  

pro_list
```

```python
# 跑apriori演算法
result = list(apriori(pro_list, min_support=0.001, min_confidence=0.2, min_lift=2, length=2)) 
result
```

```python
#support
rank_support = sorted(result, reverse = True, key = lambda RelationRecord : RelationRecord.support)
for results in rank_support:
    print(list(results.items))
    print(results.support, '\n')
```

```python
#confidence
rank_confidence = sorted(result, reverse = True, key = lambda RelationRecord : RelationRecord.ordered_statistics[0][3])
for results in rank_confidence:
    print(list(results.items))
    print(results.ordered_statistics[0][2], '\n')
```

```python
rank_lift = sorted(result, reverse = True, key = lambda RelationRecord : RelationRecord.ordered_statistics[0][3])
for results in rank_lift:
    print(list(results.items))
    print(results.ordered_statistics[0][3], '\n')
```

## 觀察結論
* 很多同品牌的東西lift值非常非常的高，可能是bundle在一起賣的，所以不太準確，取非同品牌組合比較有用。
* 由於政大比較偏遠，但在寒暑假銷售額卻沒有掉非常多，且support值高的商品裡有許多生食，推測主要客群為家庭主婦，而非政大學生。

## 推薦改善方案
雖然福利社主要目的是服務學校師生，不是很在意銷售額的多寡，但我們假設福利社有想要增加營收
* 多放廣告讓學生知道來這邊買東西比較便宜。
* 針對原有的家庭主婦客戶，利用上面的關聯分析，找出關聯性強的商品做搭配促銷或將其擺設在一起，促進客戶一起購買。
* 由於現在的文具擺設較前面，生活用品擺設較後面，在學生客戶尚未增加足夠前，可以將兩者位子調換，方便家庭主婦們優先看到生活用品，增加購買機率。
