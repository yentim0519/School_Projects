# 將youtube影片顯示在jupyter notebook上

```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import re
from IPython.display import display, YouTubeVideo
```

```python
def display_youtube_video(url_string):
    video_id = url_string.replace('https://www.youtube.com/watch?v=', '')
    myvideo = YouTubeVideo(video_id)
    display(myvideo)
```

```python
display_youtube_video('https://www.youtube.com/watch?v=yoiCSrxvwjI')
```
