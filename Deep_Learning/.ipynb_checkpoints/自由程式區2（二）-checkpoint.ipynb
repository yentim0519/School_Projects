# 將台灣百岳在地圖上顯示出來（folium)

```python
import urllib
from bs4 import BeautifulSoup
from pyproj import Proj
import folium
```

```python
request = urllib.request.urlopen('http://www.mountaineering.org.tw/tw/index.php/2015-09-13-14-48-49/2015-09-13-15-07-13').read()
soup = BeautifulSoup(request, "html.parser")
table_tags = soup.find_all('table')
```

```python
tr_list = []
for table_tag in table_tags:
    tr_list.append(table_tag.find_all('tr'))
```

```python
myProj = Proj("+proj=tmerc +ellps=aust_SA +lon_0=121 +x_0=250000 +k=0.9999 +units=m +no_defs")
```

```python
df = pd.DataFrame(columns = ['山名', '橫座標', '縱座標', '標高'])
```

```python
for i in range(len(tr_list)):
    for j in range(2, len(tr_list[i])):   
        longt, lat = myProj(float(tr_list[i][j].find_all('p')[2].string), float(tr_list[i][j].find_all('p')[3].string), inverse=True)
        df = df.append(pd.Series([tr_list[i][j].find_all('p')[1].string, longt, lat, tr_list[i][j].find_all('p')[4].string], index=df.columns), ignore_index=True)
```

```python
df
```

```python
mount_map = folium.Map([23.770876, 120.934757], zoom_start=8, tiles='Cartodb Positron')
```

```python
# 解決顯示的編碼問題
def utf2asc(s):
    return str(str(s).encode('ascii', 'xmlcharrefreplace'))[2:-1]
```

```python
for i in range(len(df)):
    folium.RegularPolygonMarker([df.iloc[i, 2], df.iloc[i, 1]],
                                 popup= f'{utf2asc(df.iloc[i, 0])}, {df.iloc[i, 3]}',
                                 number_of_sides=3,
                                 rotation=30,
                                 radius=15 ,
                                 color='green',
                                 fill_color='green',
                                 fill_opacity=0.5).add_to(mount_map)
```
